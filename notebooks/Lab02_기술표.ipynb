## 📋 Document AI 기술 스택

```
┌─────────────────────────────────────────────────┐
│           비즈니스 응용 계층                      │
│  계약서 분석 | 영수증 처리 | 신분증 인식         │
└─────────────────────────────────────────────────┘
                       ↓
┌─────────────────────────────────────────────────┐
│           정보 추출 계층 (NER)                   │
│  규칙기반 | 딥러닝 | LLM 기반                    │
└─────────────────────────────────────────────────┘
                       ↓
┌─────────────────────────────────────────────────┐
│           텍스트 인식 계층 (OCR)                 │
│  Tesseract | PaddleOCR | EasyOCR | 상용 API    │
└─────────────────────────────────────────────────┘
                       ↓
┌─────────────────────────────────────────────────┐
│           레이아웃 분석 계층                      │
│  표 검출 | 영역 분할 | 구조 분석                │
└─────────────────────────────────────────────────┘
                       ↓
┌─────────────────────────────────────────────────┐
│           전처리 계층                            │
│  노이즈 제거 | 이진화 | 기울기 보정              │
└─────────────────────────────────────────────────┘
```


## 🔍 OCR 엔진 비교표

| 엔진 | 정확도 | 속도 | 한글지원 | 비용 | 추천용도 |
|------|--------|------|----------|------|----------|
| **Tesseract** | ⭐⭐⭐ | ⚡⚡⚡ | ✅ | 무료 | 빠른 프로토타입 |
| **PaddleOCR** | ⭐⭐⭐⭐ | ⚡⚡ | ✅✅ | 무료 | 다국어 문서 |
| **EasyOCR** | ⭐⭐⭐⭐ | ⚡⚡ | ✅✅ | 무료 | 손글씨, 복잡한 레이아웃 |
| **Google Vision** | ⭐⭐⭐⭐⭐ | ⚡⚡⚡ | ✅✅✅ | $1.5/1000 | 고정확도 필요 |
| **AWS Textract** | ⭐⭐⭐⭐⭐ | ⚡⚡⚡ | ✅✅ | $1.5/1000 | 표/양식 분석 |
| **Upstage** | ⭐⭐⭐⭐⭐ | ⚡⚡⚡ | ✅✅✅ | 문의 | 한글 특화, 엔터프라이즈 |

### 선택 가이드

**무료로 시작**: Tesseract → PaddleOCR → EasyOCR 순으로 테스트

**상용 서비스**: Google Vision (범용) / AWS Textract (양식) / Upstage (한글)
